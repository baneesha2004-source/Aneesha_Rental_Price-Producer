```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "store_nbr":[1,2,3],
    "tot_sales":[200,300,400]
}
df=pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["store_nbr"], df["tot_sales"])
plt.show()
```

```python
store_sales=df.groupby("Store_Nbr")["Tot_Sales"].sum()
plt.bar(store_sales.index,store_sales.values)
plt.show()
```

```python
store_sales=df.groupby("store_nbr")["tot_sales"].sum()
plt.bar(store_sales.index,store_sales.values)
plt.show()
```

```python
df["date"]=pd.to_datetime(df["date"])
daily_sales=df.groupby("date")["tot_sales"].sum()
plt.plot(daily_sales.index,daily_sales.values)
plt.show()
```

```python
df["DATE"]=["2024-01-01","2024-01-02","2024-01-03"]
daily_sales=df.groupby("DATE")["tot_sales"].sum()
plt.plot(daily_sales.index,daily_sales.values)
plt.show()
```

```python
sales=df.groupby("store_nbr")["tot_sales"].sum()
plt.pie(sales.values, labels=sales.index, autopct="%1.1f%%")
plt.show()
```

```python
plt.hist(df["tot_sales"])
plt.show()
```

```python
plt.plot(daily_sales.index , daily_sales.values)
plt.xlabel("Date")
plt.ylabel("sales")
plt.title("daily sales trend")
plt.show()
```
